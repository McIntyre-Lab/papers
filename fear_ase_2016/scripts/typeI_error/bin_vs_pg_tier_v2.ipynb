```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## PG vs Binomial

### Interspecific

```python
# PG vs Binomail Data
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/misspecification_tier_binomial_vs_pg_plotting.csv'
inter = pd.read_csv(fname)

def inter_make_pgbin_plot(ax):
    global inter
    ax.plot(inter['per_miss'], inter['binomial'], '^-', label='Exact Binomial Test')
    ax.plot(inter['per_miss'], inter['pg'], '*--', label='PG model q=1/2')
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper left')
    ax.set_xlim(-2, 32)
    ax.set_ylim(0,1)
    ax.set_xlabel('Percentage of Misspecification (from 0.5)', fontsize=14)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
    
inter[:3]
```

### Intraspecific

```python
# PG vs Binomail Data
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/r101_misspecification_tier_binomial_vs_pg_plotting.csv'
intra = pd.read_csv(fname)

import matplotlib.ticker as ticker

def make_patch_spines_invisible(ax):
    ax.set_frame_on(True)
    ax.patch.set_visible(False)
    for sp in ax.spines.itervalues():
        sp.set_visible(False)
        
def make_spine_invisible(ax, direction):
    if direction in ['right', 'left']:
        ax.yaxis.set_ticks_position(direction)
        ax.yaxis.set_label_position(direction)
    elif direction in ['top', 'bottom']:
        ax.xaxis.set_ticks_position(direction)
        ax.xaxis.set_label_position(direction)
    else:
        raise ValueError("Unknown Direction : %s" % (direction,))
    
    ax.spines[direction].set_visible(True)


def intra_make_pgbin_plot(ax):
    global intra
    ax.plot(intra['per_miss'], intra['binomial'], '^-', label='Exact Binomial Test')
    ax.plot(intra['per_miss'], intra['pg4'], 'o-', label='PG model q=0.4')
    ax.plot(intra['per_miss'], intra['pg5'], '*--', label='PG model q=0.5')
    ax.plot(intra['per_miss'], intra['pg6'], 'D:', label='PG model q=0.6')
    ax.plot(intra['per_miss'], intra['pgemp'], '-', label='Intersection q= {0.4, 0.5, 0.6}', color='purple', lw=5, alpha=0.3)
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper middle')
    ax.set_xlim(-32, 32)
    ax.set_ylim(0,1)
    ax.set_xlabel('Percentage of Misspecification (from 0.5)', fontsize=11)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
    
    # Set x-axis tick marks
    ax.xaxis.set_major_locator(ticker.FixedLocator([-30, -20, -10, 0, 10, 20, 30]))
    
    plt.subplots_adjust(bottom=0.12) # Make rooom for additional axes
    
    # Create a Second x-axis for 0.4
    ax2 = ax.twiny()
    ax2.spines["bottom"].set_position(("axes", -.16))
    make_patch_spines_invisible(ax2)
    make_spine_invisible(ax2, "bottom")
    ax2.set_xlabel("Percentage of Misspecification (from 0.4)", fontsize=11)
    ax2.set_xlim(-32, 32)
    ax2.xaxis.set_major_locator(ticker.FixedLocator([-30, -20, -10, 0, 10, 20, 30]))
    ax2.xaxis.set_ticklabels([-10, 0, 10, 20, 30, 40, 50])

    # Create a Second x-axis for 0.6
    ax3 = ax.twiny()
    ax3.spines["bottom"].set_position(("axes", -.32))
    make_patch_spines_invisible(ax3)
    make_spine_invisible(ax3, "bottom")
    ax3.set_xlabel("Percentage of Misspecification (from 0.6)", fontsize=11)
    ax3.set_xlim(-32, 32)
    ax3.xaxis.set_major_locator(ticker.FixedLocator([-30, -20, -10, 0, 10, 20, 30]))
    ax3.xaxis.set_ticklabels([-50, -40, -30, -20, -10, 0, 10])
 
    

    
intra[:3]
```

### Make Plot

```python
fig = plt.figure()
fig.set_size_inches(12,6)
for i, label in enumerate(('A', 'B')):
    ax = fig.add_subplot(1,2,i+1)
    ax.text(-0.07, 1.03, label, transform=ax.transAxes,
      fontsize=16, fontweight='bold', va='top', ha='right')
    if i == 0:
        inter_make_pgbin_plot(ax)
    else:
        intra_make_pgbin_plot(ax)

plt.tight_layout()
fig.savefig('/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/tier_bin_vs_pg.png')
```
