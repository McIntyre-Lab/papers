```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## PG vs Binomial

### Interspecific

```python
#Misspecification Data
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/misspecification_bias_for_plotting.csv'
inter = pd.read_csv(fname)

def inter_make_miss_plot(ax):
    global inter
    
    ax.plot(inter['sim_bias'], inter['tier_00'], 's-', label = '0%')
    ax.plot(inter['sim_bias'], inter['tier_01'], 'o--', markersize=3, label = '1%')
    ax.plot(inter['sim_bias'], inter['tier_02'], '^:', label = '2%')
    ax.plot(inter['sim_bias'], inter['tier_05'], 'D-.', label = '5%')
    ax.plot(inter['sim_bias'], inter['tier_10'], 'o-', label = '10%')
    ax.plot(inter['sim_bias'], inter['tier_20'], '+--', label = '20%')
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper left')
    ax.set_xlim(-35,35)
    ax.set_ylim(0,1)
    ax.set_xlabel('Simulated Amount of Bias', fontsize=14)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
    
inter[:3]
```

### Intraspecific

```python
#Misspecification Data
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/r101_misspecification_bias_for_plotting.csv'
intra = pd.read_csv(fname)

def intra_make_miss_plot(ax):
    global intra
    
    ax.plot(intra['sim_bias'], intra['tier_00'], 's-', label = '0%')
    ax.plot(intra['sim_bias'], intra['tier_01'], 'o--', markersize=3, label = '1%')
    ax.plot(intra['sim_bias'], intra['tier_02'], '^:', label = '2%')
    ax.plot(intra['sim_bias'], intra['tier_05'], 'D-.', label = '5%')
    ax.plot(intra['sim_bias'], intra['tier_10'], 'o-', label = '10%')
    ax.plot(intra['sim_bias'], intra['tier_20'], '+--', label = '20%')
    ax.plot(intra['sim_bias'], intra['tier_emp'], '-', label = 'Intersection q = {0.4, 0.5, 0.6}', color='k')
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper left')
    ax.set_xlim(-35,35)
    ax.set_ylim(0,1)
    ax.set_xlabel('Simulated Amount of Bias', fontsize=14)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
    
intra[:3]
```

### Make Plot

```python
fig = plt.figure()
fig.set_size_inches(12,6)
for i, label in enumerate(('A', 'B')):
    ax = fig.add_subplot(1,2,i+1)
    ax.text(-0.07, 1.03, label, transform=ax.transAxes,
      fontsize=16, fontweight='bold', va='top', ha='right')
    if i == 0:
        inter_make_miss_plot(ax)
    else:
        intra_make_miss_plot(ax)

plt.tight_layout()
fig.savefig('/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/tier_misspecification_vs_bias.png')
```
