```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## PG vs Binomial

### Interspecific

```python
# PG vs Binomail Data
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/misspecification_tier_binomial_vs_pg_plotting.csv'
inter = pd.read_csv(fname)

def inter_make_pgbin_plot(ax):
    global inter
    ax.plot(inter['per_miss'], inter['binomial'], '^-', label='Exact Binomial Test')
    ax.plot(inter['per_miss'], inter['pg'], '*--', label='PG model q=1/2')
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper left')
    ax.set_xlim(-2, 32)
    ax.set_ylim(0,1)
    ax.set_xlabel('Percentage of Misspecification', fontsize=14)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
    
inter[:3]
```

### Intraspecific

```python
# PG vs Binomail Data
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/r101_misspecification_tier_binomial_vs_pg_plotting.csv'
intra = pd.read_csv(fname)

def intra_make_pgbin_plot(ax):
    global intra
    ax.plot(intra['per_miss'], intra['binomial'], '^-', label='Exact Binomial Test')
    ax.plot(intra['per_miss'], intra['pg4'], 'o-', label='PG model q=0.4')
    ax.plot(intra['per_miss'], intra['pg5'], '*--', label='PG model q=0.5')
    ax.plot(intra['per_miss'], intra['pg6'], 'D:', label='PG model q=0.6')
    ax.plot(intra['per_miss'], intra['pgemp'], '-', label='Intersection q= {0.4, 0.5, 0.6}', color='purple', lw=5, alpha=0.3)
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper middle')
    ax.set_xlim(-32, 32)
    ax.set_ylim(0,1)
    ax.set_xlabel('Percentage of Misspecification', fontsize=14)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
    
intra[:3]
```

### Make Plot

```python
fig = plt.figure()
fig.set_size_inches(12,6)
for i, label in enumerate(('A', 'B')):
    ax = fig.add_subplot(1,2,i+1)
    ax.text(-0.07, 1.03, label, transform=ax.transAxes,
      fontsize=16, fontweight='bold', va='top', ha='right')
    if i == 0:
        inter_make_pgbin_plot(ax)
    else:
        intra_make_pgbin_plot(ax)

plt.tight_layout()
fig.savefig('/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/tier_bin_vs_pg.png')
```
