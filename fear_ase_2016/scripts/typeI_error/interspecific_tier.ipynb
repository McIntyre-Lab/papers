```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## PG vs Binomial

```python
# PG vs Binomail Data
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/misspecification_tier_binomial_vs_pg_plotting.csv'
pgbin = pd.read_csv(fname)

def make_pgbin_plot(ax):
    global pgbin
    ax.plot(pgbin['per_miss'], pgbin['binomial'], '^-', label='Exact Binomial Test')
    ax.plot(pgbin['per_miss'], pgbin['pg'], '*--', label='PG model q=1/2')
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper left')
    ax.set_xlim(-2, 32)
    ax.set_xlabel('Percentage of Misspecification', fontsize=14)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
```

```python
#Misspecification Data
fname = '/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/misspecification_bias_for_plotting.csv'
miss = pd.read_csv(fname)

def make_miss_plot(ax):
    global miss
    
    ax.plot(miss['sim_bias'], miss['tier_00'], 's-', label = '0%')
    ax.plot(miss['sim_bias'], miss['tier_01'], 'o--', markersize=3, label = '1%')
    ax.plot(miss['sim_bias'], miss['tier_02'], '^:', label = '2%')
    ax.plot(miss['sim_bias'], miss['tier_05'], 'D-.', label = '5%')
    ax.plot(miss['sim_bias'], miss['tier_10'], 'o-', label = '10%')
    ax.plot(miss['sim_bias'], miss['tier_20'], '+--', label = '20%')
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper left')
    ax.set_xlim(-35,35)
    ax.set_xlabel('Simulated Amount of Bias', fontsize=14)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
```

```python
fig = plt.figure()
fig.set_size_inches(12,6)
for i, label in enumerate(('A', 'B')):
    ax = fig.add_subplot(1,2,i+1)
    ax.text(-0.07, 1.03, label, transform=ax.transAxes,
      fontsize=16, fontweight='bold', va='top', ha='right')
    if i == 0:
        make_pgbin_plot(ax)
    else:
        make_miss_plot(ax)

plt.tight_layout()
fig.savefig('/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/output/interspecific_tier.png')
```
