```python
def intra_make_miss_plot(df, ax, title):  
    ax.set_title(title)
    ax.plot(df['sim_bias'], df['tier_00'], 's-', label = '0%')
    ax.plot(df['sim_bias'], df['tier_01'], 'o--', markersize=3, label = '1%')
    ax.plot(df['sim_bias'], df['tier_02'], '^:', label = '2%')
    ax.plot(df['sim_bias'], df['tier_05'], 'D-.', label = '5%')
    ax.plot(df['sim_bias'], df['tier_10'], 'o-', label = '10%')
    ax.plot(df['sim_bias'], df['tier_20'], '+--', label = '20%')
    ax.plot(df['sim_bias'], df['tier_emp'], '-', label = 'Intersection q = {0.4, 0.5, 0.6}', color='k')
    ax.axhline(0.05, ls='--', lw=2, color='red', label='TIER 0.05', alpha=.4)
    ax.legend(loc='upper left')
    ax.set_xlim(-35,35)
    ax.set_ylim(0,0.4)
    ax.set_xlabel('Simulated Amount of Bias', fontsize=14)
    ax.set_ylabel('Type I Error Rate', fontsize=14)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))
intra_make_miss_plot(dfr361, ax1, 'r361 Low Coverage')
intra_make_miss_plot(dfr332, ax2, 'r332 Medium Coverage')
intra_make_miss_plot(dfr365, ax3, 'r365 High Coverage')
#intra_make_miss_plot(dfr101, ax4, 'r101 FB551')
plt.tight_layout()
fig.savefig('/home/jfear/mclab/cegs_ase_paper/pipeline_output/typeI_error/tier_misspecification.png')
```
