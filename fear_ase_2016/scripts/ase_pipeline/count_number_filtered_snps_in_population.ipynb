# Count Number Filtered SNPs
Count the number of SNPs for each genotype after level 2 filtering.

```python
import re
import glob
import gzip
from collections import defaultdict

pjoin = os.path.join
```

```python
# Level 2 filtered VCF files can be downloaded from:
# http://bio.rc.ufl.edu/pub/mcintyre/papers/fear_ase_2015/ase_lvl2_filtered_vcf_files/
# OR UF-Hipergator users at:
# /bio/mcintyre/papers/fear_ase_2015/ase_lvl2_filtered_vcf_files/
VCF_PATH = os.path.expanduser('~/sandbox/ase_lvl2_filtered_vcf_files')
```

```python
cnts = defaultdict(dict)
for f in glob.glob(pjoin(VCF_PATH, '*.gz')):
    # Pull apart the file name
    name, ref = os.path.basename(f).split('_')[:2]
    
    # If name is w1118 then call tester
    if name == 'w1118':
        name = 'tester'
    else:
        name = 'line'
        
    # Figure out which ref it is
    ref = re.sub('w11182', '', ref)
    
    # Count number of SNPs/InDels
    with gzip.open(f, 'rb') as FH:
        cnt = sum(1 for _ in FH)
    
    cnts[name][ref] = cnt    
```
